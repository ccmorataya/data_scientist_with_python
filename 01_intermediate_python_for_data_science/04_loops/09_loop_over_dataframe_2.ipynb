## Loop over DataFrame (2)

The row data that's generated by [<u>`iterrows()`</u>](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.iterrows.html) on every run is a Pandas Series. This format is not very convenient to print out. Luckily, you can easily select variables from the Pandas Series using square brackets:

``` python
for lab, row in brics.iterrows() :
    print(row['country'])
```
